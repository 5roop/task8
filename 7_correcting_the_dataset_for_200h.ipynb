```python
import pandas as pd
df = pd.read_csv("keep_pruned.csv")
df["duration"] = df.end - df.start
df["cumulative_duration"] = df.duration.cumsum()
df.head()
```

```python
c = df.cumulative_duration < 200*3600
df.loc[c, :].to_csv("keep_pruned.csv")
```

```python
from concurrent.futures import ProcessPoolExecutor
import os
def transform(file):
    os.system(f"ffmpeg -i {file} -ac 1 {file}")
with ProcessPoolExecutor(max_workers=20) as executor:
    executor.map(transform, ["transfer/" + i for i in df.loc[~c, "hashname"].values])
```
